```python
# Import required libraries
from scipy import special
import math
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Data from the table
omega_L=[0.1,0.14,0.17,0.20,0.22,0.23]
delta_t_cs=[15,32.01,51.01,65.51,80.01,88.01]
delta_t_cl=[75,92.01,111.01,125.51,140.01,148.01]

# Create scatter plot
plt.scatter(omega_L,delta_t_cs,marker='s',color='orange',label=r'$\Delta t_{cs}$')
plt.scatter(omega_L,delta_t_cl,marker='o',color='blue',label=r'$\Delta t_{cl}$')

# Axis labels
xlabel_size=16
ylabel_size=16
plt.xlabel(r'$\omega_L$ (fs$^{-1}$)',fontsize=xlabel_size)
plt.ylabel('Time (fs)',fontsize=ylabel_size)

# Axis formatting
ax=plt.gca()
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f'))
ax.tick_params(axis='both',direction='in',which='both',bottom=True,top=True,left=True,right=True)

# Increase tick label size
tick_label_size=14
ax.tick_params(axis='both',labelsize=tick_label_size)

# Add legend and display plot
plt.legend()
plt.show()
```
